**Import the Packages**

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snspi 
```

**Read the data**

```python
data = pd.read_table('Nichols_Nicki.dat')
data
```

**Successfully remove the two variables with low variances**

```python
stats = data.describe().T.sort_values("std").head()
print(stats)
```

**Alternative way to remove the two variables with low variances**

```python
variances = data.var()
lowest_variance_items = variances.nsmallest(2).index.tolist()
print("Items dropped for low variances:", lowest_variance_items)
var_drop = data.drop(['ias34','ias23'], axis= 1)
var_drop
```

```python
var_drop.corr()
```

**Look at the confusion matrix and remove the three variables with low correlations**

```python
corr = var_drop.corr()

# # # If you want to see the matrix in a csv, uncomment below
# conf_df = pd.DataFrame(corr)
# conf_df.to_csv('confusion_matrix.csv', index=True, header=False)

f_data = var_drop.drop(['ias13', 'ias22', 'ias32'], axis=1)
f_data.corr()
```

**Correct the eigenvalues**

```python
#Calculating the eigen value
# f_cor = f_data.corr()
cov_mat = np.cov(f_data.T)
#np.cov(f_data)

eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
eigen_vals
```

**PCA with all n_components**

```python
pca = PCA(n_components=None)
f_data_pca = pca.fit_transform(f_data)
pca.explained_variance_ratio_
```

**PCA with 5 n_components**

```python
pca_5 = PCA(n_components=5)
f_data_pca2 = pca_5.fit_transform(f_data)
pca_5.explained_variance_ratio_
```
