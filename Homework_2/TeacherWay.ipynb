```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snspi 
```

```python
data = pd.read_table('Nichols_Nicki.dat')
data
```

```python
stats= data.describe().T.sort_values("std").head()
print(stats)
data2 = data.drop(['ias34','ias23'], axis=1)
print(data2)
```

```python
cov_mat = np.cov(data2.T)
print(cov_mat)

# conf_df = pd.DataFrame(cov_mat)
# conf_df.to_csv('hcm.csv', index=False, header=False)

data3 = data2.drop(['ias13', 'ias22', 'ias32'],axis=1)
data3
```

```python
cov_mat2 = np.cov(data3.T)
cov_mat2

eigen_vals, eigen_vecs = np.linalg.eig(cov_mat2)
eigen_vals
```

```python
pca = PCA(n_components=None)
f_data_pca = pca.fit_transform(data3)
pca.explained_variance_ratio_
```
