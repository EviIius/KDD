```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = 'MovieReviews' # Folder Name
filelists = PlaintextCorpusReader(corpus_root, '.*',encoding='latin-1')  # wildcard is read all files in the folder
filelists.fileids()  # Get the filenames
```

```python
#create a list of news articles
news = []
for fileid in filelists.fileids():
    news.append(filelists.raw(fileid))
```

```python
news[25]
```

```python
# import the shorttext library for text preprocessing
#standard_text_preprocessor_1 under shorttext.utils provides a standar way of text preprocessing, including the following steps:

   #1. removing special characters,
   #2. removing numerals,
   #3. converting all alphabets to lower cases,
   #4. removing stop words, and
   #5. stemming the words (using Porter stemmer).

from shorttext.utils import standard_text_preprocessor_1, DocumentTermMatrix
preprocessor = standard_text_preprocessor_1()
corpus = [preprocessor(article).split(' ') for article in news]
```

```python
corpus[25]
```

```python
#convert the corpus of news to a document term matrix (dtm), where each row represents a document, each column represents the number of occurence for a word

dtm = DocumentTermMatrix(corpus, docids = filelists.fileids())
```

```python
# define a function to calculate entropy given a distribution
import numpy as np
import math

def entropy(p):
    if sum(p) == 0:
        return 0

    p = p/sum(p)

    p = p[ p > 0 ]

    H = -sum(p*np.log2(p))

    return H
```

```python
from collections import Counter

# Create a list of authors
authors = ['Berardinelli']*80 + ['Schwartz']*100

# Define a function to calculate mutual information
def mutual_information(word, authors, dtm):
    word_occurrences = dtm.get_token_occurences(word)
    author_word_counts = Counter([authors[i] for i in range(len(authors)) if word in corpus[i]])
    total_reviews = len(authors)
    total_word_occurrences = sum(word_occurrences.values())
    
    mi = 0
    for author in set(authors):
        for word_presence in [0, 1]:
            # Calculate joint probability
            if word_presence == 1:
                joint_prob = author_word_counts[author] / total_reviews
            else:
                joint_prob = (total_word_occurrences - author_word_counts[author]) / total_reviews
            
            # Calculate marginal probabilities
            prob_word = total_word_occurrences / total_reviews
            prob_author = authors.count(author) / total_reviews
            
            if joint_prob > 0 and prob_word > 0 and prob_author > 0:
                mi += joint_prob * np.log2(joint_prob / (prob_word * prob_author))
    return mi
```

```python
# Calculate the mutual information for the word "director"
mi_director = mutual_information('director', authors, dtm)
print(f'Mutual Information of "director": {mi_director}')
```

```python
vocabulary = {}

for document in corpus:
    for word in document:
        if word not in vocabulary:
            vocabulary[word] = 1
        else:
            vocabulary[word] += 1
```

```python
# Create an empty dictionary to store the mutual information values for each word
mi_all_words = {}

# Iterate over the vocabulary
for word in vocabulary.keys():
    # Calculate the mutual information for the word
    mi = mutual_information(word, authors, dtm)
    # Store the mutual information value in the dictionary
    mi_all_words[word] = mi

# Sort the words based on their mutual information values in descending order
sorted_words = sorted(mi_all_words.items(), key=lambda x: x[1], reverse=True)

# Get the top ten words
top_ten_words = sorted_words[:10]

# Print the top ten words
for word, mi in top_ten_words:
    print(f'Word: {word}, Mutual Information: {mi}')
```
