```python
import pandas as pd
import numpy as np
import plotly as p
import mlxtend as m 
```

```python
dataset = pd.read_csv("Titanic.csv")
dataset.shape
```

```python
dataset
```

```python
dataset.describe()
```

```python
# Gather All Items of Each Transactions into Numpy Array
transaction = []
for i in range(0, dataset.shape[0]):
    for j in range(0, dataset.shape[1]):
        transaction.append(dataset.values[i,j])
# converting to numpy array
transaction = np.array(transaction)
print(transaction)
```

```python
#  Transform Them a Pandas DataFrame
df = pd.DataFrame(transaction, columns=["items"])
# Put 1 to Each Item For Making Countable Table, to be able to perform Group By
df["incident_count"] = 1
df
```

```python
#  Delete NaN Items from Dataset
indexNames = df[df['items'] == "nan" ].index
df.drop(indexNames , inplace=True)
# Making a New Appropriate Pandas DataFrame for Visualizations
df_table = df.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
df_table
```

```python
# Transform Every Transaction to Seperate List & Gather Them into Numpy Array
transaction2 = []
for i in range(dataset.shape[0]):
    transaction2.append([str(dataset.values[i,j]) for j in range(dataset.shape[1])])
```

```python
type(transaction2)
```

```python
transaction2
```

```python
transaction2 = np.array(transaction2)
```

```python
transaction2
```

```python
# importing the required module
from mlxtend.preprocessing import TransactionEncoder
# initializing the transactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transaction2).transform(transaction2)
```

```python
te_ary
```

```python
dataset2 = pd.DataFrame(te_ary, columns=te.columns_)
# dataset after encoded
dataset2
```

```python
# select top 30 items
first30 = df_table["items"].head(35).values
# Extract Top 30
dataset3 = dataset2.loc[:,first30]
# shape of the dataset
dataset3.shape
```

```python
#Importing Libraries
from mlxtend.frequent_patterns import fpgrowth
#running the fpgrowth algorithm
res=fpgrowth(dataset3,min_support=0.1, use_colnames=True)
res.head(35)
```

```python
# importing required module
from mlxtend.frequent_patterns import association_rules
# creating asssociation rules
rules=association_rules(res, metric="confidence", min_threshold=.7)
# printing association rules
rules
```

```python
# Filter rules where the consequent is a value of the "Survived" variable
survived_rules = rules[rules['consequents'].apply(lambda x: len(x) == 1 and ('Yes' in x or 'No' in x))]
survived_rules
```
